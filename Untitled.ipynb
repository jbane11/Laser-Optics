```python

def angleperstep(DF,anglearray):
    trial = DF.copy()
    trial.loc[trial['LDD1 Current(A)'] > 21.0, 'Laser status'] = 1
    trial.loc[trial['LDD1 Current(A)'] <= 21.0, 'Laser status'] = 0
    trial.loc[trial['Sensor Power'] <= 0.003, 'Sensor status'] = int(0)
    trial.loc[trial['Sensor Power'] > 0.003, 'Sensor status'] = int(1)
    steparray =[]
    step=0
    laststep=0
    lazeonlen=0
    lazeonarray=[]
    dataQ=[]
    for i,j in enumerate(trial.index):
        if i ==0 :
            steparray.append(step)
            lazeonarray.append(lazeonlen)
            dataQ.append(1)
            continue
        if j not in trial.index:
            print("Index issue")
        

        row=trial.iloc[i].copy()
        
        #print(row)
        lastrow=trial["Laser status"].iloc[i-1]
        try :
            nextrow=trial["Laser status"].iloc[i+1]
        except:
            nextrow=0
        las=row["Laser status"]
        sens=row["Sensor status"]
     
        if las == 1 and lastrow ==0 and nextrow ==0:
            las=0
            trial["Laser status"].iloc[i]=0
                
        if las != lastrow :
            if lazeonlen <= 20 and las ==0 and step !=0 and lazeonlen !=0 :
                print("shrt run correction" ,  "i",i,"j",j,"dtep",step,"len:",lazeonlen) 
                print(lazeonlen,las,step)
                dataQ.append(0)
            else:
                dataQ.append(1)
          
            if las == 1:
                
                step=step+1
                steparray.append(step)
                lazeonarray.append(lazeonlen)
                
                print("new Step",  i,j,step, j-laststep,lazeonlen)
                lazeonlen=0
                laststep=j
            else:
                lazeonarray.append(0)
                steparray.append(step)
                #lazeonarray.append(lazeonlen)
            
        else:
            dataQ.append(1)
            if las ==1:
                lazeonlen=lazeonlen+1
                lazeonarray.append(lazeonlen)
            else:
                lazeonarray.append(0)
                lazeonlen=0
            steparray.append(step)
        
    if max(steparray) != len(anglearray):
            print("issue", max(steparray),len(anglearray))
           # print(steparray)        
    else:
        print(max(steparray),anglearray)
       # print(steparray)   
        trial["Angle"]=anglearray[steparray]
    
    trial["step"]=steparray
    trial["dataQ"]=dataQ
    trial["LaserOnlen"]=lazeonarray
    return trial
```

```python


def angleperstep(DF,anglearray):
    trial = DF.copy()
    trial.loc[trial['LDD1 Current(A)'] > 21.0, 'Laser status'] = 1
    trial.loc[trial['LDD1 Current(A)'] <= 21.0, 'Laser status'] = 0
    trial.loc[trial['Sensor Power'] <= 0.003, 'Sensor status'] = int(0)
    trial.loc[trial['Sensor Power'] > 0.003, 'Sensor status'] = int(1)
    steparray =[]
    step=0
    laststep=0
    lazeonlen=0
    lazeonarray=[]
    dataQ=[]
    for i,j in enumerate(trial.index):
        if i ==0 :
            steparray.append(step)
            lazeonarray.append(lazeonlen)
            dataQ.append(1)
            continue
        if j not in trial.index:
            print("Index issue")
        

        row=trial.iloc[i].copy()
        
        #print(row)
        lastrow=trial["Laser status"].iloc[i-1]
        try :
            nextrow=trial["Laser status"].iloc[i+1]
        except:
            nextrow=0
        las=row["Laser status"]
        sens=row["Sensor status"]
     
        if las == 1 and lastrow ==0 and nextrow ==0:
            las=0
            trial["Laser status"].iloc[i]=0
                
        if las != lastrow :
            if lazeonlen <= 20 and las ==0 and step !=0 and lazeonlen !=0 :
                print("shrt run correction" ,  "i",i,"j",j,"dtep",step,"len:",lazeonlen) 
                print(lazeonlen,las,step)
                dataQ.append(0)
            else:
                dataQ.append(1)
          
            if las == 1:
                
                step=step+1
                steparray.append(step)
                lazeonarray.append(lazeonlen)
                
                #print("new Step",  i,j,step, j-laststep,lazeonlen)
                lazeonlen=0
                laststep=j
            else:
                lazeonarray.append(0)
                steparray.append(step)
                #lazeonarray.append(lazeonlen)
            
        else:
            dataQ.append(1)
            if las ==1:
                lazeonlen=lazeonlen+1
                lazeonarray.append(lazeonlen)
            else:
                lazeonarray.append(0)
                lazeonlen=0
            steparray.append(step)
        
    if max(steparray) != len(anglearray)-1:
            print("issue", max(steparray),len(anglearray))
           # print(steparray)        
    else:
        print(max(steparray),anglearray)
        print(type(steparray[1]))
        angle_list= [float(anglearray[k]) for k in steparray]
            
        trial["Angle"]=angle_list
    
    trial["step"]=steparray
    trial["dataQ"]=dataQ
    trial["LaserOnlen"]=lazeonarray
    return trial



def getanglepowerstd(DF,angles):
    angle_power_std=[]
    for angle in angles[1:]:
        DF_anglecut = DF.query("Angle == %f and `Laser status` == 1 and LaserOnlen > 3"%(float(angle)))
        Full_angle_power = np.array(DF_anglecut["Sensor Power"])
        
        
       # print(angle,len(DF_anglecut)) 
        
        if len(Full_angle_power) > 0:  
            Full_angle_avg = Full_angle_power.sum()/len(Full_angle_power)
            #print(max(Full_angle_power),min(Full_angle_power))
            max_sep = 1* max( [0,abs(Full_angle_avg-max(Full_angle_power)), abs(Full_angle_avg-min(Full_angle_power))])
            #print(max_sep)
            angle_power = np.array(DF_anglecut.query("abs(`Sensor Power` - %f) < %f"%(Full_angle_avg,max_sep))["Sensor Power"])
            angle_power_std.append( np.std(angle_power))
            
            sum_of_diff = np.abs(angle_power)
            
        else:
            print("issue with no events in cal")
            angle_power_std.append(0)
        #print(angle,len(DF_anglecut),np.std(angle_power))    
        #print("len",len(angle_power_std), angle,len(DF_anglecut))
    print(len(angle_power_std))
    return angle_power_std
```
